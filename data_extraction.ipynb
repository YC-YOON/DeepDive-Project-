# Load working dataset and extract subset to use for debugging
Downloaded from https://data.ca.gov/dataset/local-area-unemployment-statistics-laus/resource/b4bc4656-7866-420f-8d87-4eda4c9996ed

```python
def get_csv_file(f, **kwargs):
  loc = 'https://drive.google.com/uc?export=download&id='+f.split('/')[-2]
  out = pd.read_csv(loc, **kwargs)
  return out

def get_pkl_file(f, **kwargs):
  loc = 'https://drive.google.com/uc?export=download&id='+f.split('/')[-2]
  out = pd.read_pickle(loc, **kwargs)
  return out
```

# Convert datasets to pandas

```python
working_data = get_csv_file('https://drive.google.com/file/d/11T8XgTiHBM_esFE09oDEXHb8XKPZnAUX/view?usp=drive_link')
working_data.head()
```

```python
debugging_data = get_csv_file('https://drive.google.com/file/d/1--VemCZWUQ73s_SNkCuL-uPGDzmCi0gi/view?usp=drive_link')
debugging_data.head()
```

# Convert `Date_Numeric` column to pandas timestamps

```python
working_data['Date_Numeric'] = pd.to_datetime(working_data['Date_Numeric'], format='%m/%Y')
working_data['Date_Numeric'] = working_data['Date_Numeric'].apply(pd.Timestamp)
working_data.head()
```

```python
debugging_data['Date_Numeric'] = pd.to_datetime(debugging_data['Date_Numeric'], format='%m/%Y')
debugging_data['Date_Numeric'] = debugging_data['Date_Numeric'].apply(pd.Timestamp)
debugging_data.head()
```
